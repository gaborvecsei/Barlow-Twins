```python
import matplotlib.pyplot as plt
import numpy as np

import barlow_twins
```

```python
dataset, nb_images = barlow_twins.create_dataset("/data", 224, 224, 1, 0.6, 1.0, 1)
```

```python
images = []

for i, data in enumerate(dataset.take(10)):
    images_1 = data[0]
    images_2 = data[1]
    
    img = np.hstack((images_1[0, :, :, :].numpy(), images_2[0, :, :, :].numpy())).astype(np.uint8)
    images.append(img)
```

```python
for img in images:
    plt.figure()
    plt.imshow(img)
    plt.axis("off")
```
